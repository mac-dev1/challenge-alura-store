### Importación de datos



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

# 1. Análisis de facturación



```python
ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print(f'El ingreso de la tienda 1 es {ingreso_tienda1}')
print(f'El ingreso de la tienda 2 es {ingreso_tienda2}')
print(f'El ingreso de la tienda 3 es {ingreso_tienda3}')
print(f'El ingreso de la tienda 4 es {ingreso_tienda4}')
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda1 = dict(tienda1.groupby('Categoría del Producto',)['Producto'].count().sort_values(ascending=False))
ventas_categoria_tienda2 = dict(tienda2.groupby('Categoría del Producto',)['Producto'].count().sort_values(ascending=False))
ventas_categoria_tienda3 = dict(tienda3.groupby('Categoría del Producto',)['Producto'].count().sort_values(ascending=False))
ventas_categoria_tienda4 = dict(tienda4.groupby('Categoría del Producto',)['Producto'].count().sort_values(ascending=False))
```

```python
cat_mas_populares_tienda1 = list(ventas_categoria_tienda1.items())[:3]
cat_mas_populares_tienda2 = list(ventas_categoria_tienda2.items())[:3]
cat_mas_populares_tienda3 = list(ventas_categoria_tienda3.items())[:3]
cat_mas_populares_tienda4 = list(ventas_categoria_tienda4.items())[:3]
print(f'Las categorias mas populades de la tienda 1 son {cat_mas_populares_tienda1}')
print(f'Las categorias mas populades de la tienda 2 son {cat_mas_populares_tienda2}')
print(f'Las categorias mas populades de la tienda 3 son {cat_mas_populares_tienda3}')
print(f'Las categorias mas populades de la tienda 4 son {cat_mas_populares_tienda4}')
```

# 3. Calificación promedio de la tienda


```python
val_media_tienda1 = tienda1['Calificación'].mean()
val_media_tienda2 = tienda2['Calificación'].mean()
val_media_tienda3 = tienda3['Calificación'].mean()
val_media_tienda4 = tienda4['Calificación'].mean()
```

```python
print(f'La valoración media de la tienda 1 es {val_media_tienda1}')
print(f'La valoración media de la tienda 2 es {val_media_tienda2}')
print(f'La valoración media de la tienda 3 es {val_media_tienda3}')
print(f'La valoración media de la tienda 4 es {val_media_tienda4}')
```

# 4. Productos más y menos vendidos

```python
ventas_tienda1 = dict(tienda1.groupby('Producto')['Producto'].count().sort_values(ascending=False))
ventas_tienda2 = dict(tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False))
ventas_tienda3 = dict(tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False))
ventas_tienda4 = dict(tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False))
```

```python
mas_vendidos_tienda1 = list(ventas_tienda1.items())[:3]
mas_vendidos_tienda2 = list(ventas_tienda2.items())[:3]
mas_vendidos_tienda3 = list(ventas_tienda3.items())[:3]
mas_vendidos_tienda4 = list(ventas_tienda4.items())[:3]
```

```python
print(f'Los productos mas vendidos de la tienda 1 son {[x for x,y in mas_vendidos_tienda1]}')
print(f'Los productos mas vendidos de la tienda 2 son {[x for x,y in mas_vendidos_tienda2]}')
print(f'Los productos mas vendidos de la tienda 3 son {[x for x,y in mas_vendidos_tienda3]}')
print(f'Los productos mas vendidos de la tienda 4 son {[x for x,y in mas_vendidos_tienda4]}')
```

```python
i = 1
for tienda in [mas_vendidos_tienda1,mas_vendidos_tienda2,mas_vendidos_tienda3,mas_vendidos_tienda4]:
    x = [x for x,y in tienda]
    y = [y for x,y in tienda]
    plt.bar(x,y)
    plt.title(f'Tienda {i}')
    i += 1
    plt.show()
```

# 5. Envío promedio por tienda

```python
cost_envio_promedio_tienda1 = tienda1['Costo de envío'].mean()
cost_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
cost_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
cost_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()
```

```python
print(f'El costo de envio promedio de la tienda 1 es {cost_envio_promedio_tienda1:.2f}')
print(f'El costo de envio promedio de la tienda 2 es {cost_envio_promedio_tienda2:.2f}')
print(f'El costo de envio promedio de la tienda 3 es {cost_envio_promedio_tienda3:.2f}')
print(f'El costo de envio promedio de la tienda 4 es {cost_envio_promedio_tienda4:.2f}')
```

# 6. Gráficos

```python
plt.bar(height=[ingreso_tienda1,ingreso_tienda2,ingreso_tienda3,ingreso_tienda4],
        x=['Tienda 1','Tienda 2','Tienda 3','Tienda 4'])
plt.title('Ventas por tienda')
plt.show()
```

```python
media_categorias_populares_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].mean()
media_categorias_populares_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].mean()
media_categorias_populares_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].mean()
media_categorias_populares_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].mean()
```

```python
index = np.arange(len(media_categorias_populares_tienda1.index)) # Para obtener un indice numerico por cada categoría
# El primer parámetro de plt.barh es la altura, entonces desplazo las barras centrales la mitad de su espesor (0.1) para que queden centradas
# respecto a la categoria, y las barras externas las desplazo la misma cantidad mas su espesor (0.3), todos los desplazamientos en los 
#respectivos sentidos
plt.barh(index+0.3,media_categorias_populares_tienda1.values,label='Tienda 1',height=0.2)
plt.barh(index+0.1,media_categorias_populares_tienda2.values,label='Tienda 2',height=0.2)
plt.barh(index-0.1,media_categorias_populares_tienda3.values,label='Tienda 3',height=0.2)
plt.barh(index-0.3,media_categorias_populares_tienda4.values,label='Tienda 4',height=0.2)
plt.yticks(index,media_categorias_populares_tienda1.index)
plt.legend()
plt.title('Precios promedio por categoria')
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
plt.pie([cost_envio_promedio_tienda1,cost_envio_promedio_tienda2,cost_envio_promedio_tienda3,cost_envio_promedio_tienda4],
        labels=['Tienda 1','Tienda 2','Tienda 3','Tienda 4'],
        autopct='%1.1f%%')
plt.title('Costos de envio promedio por tienda')
plt.show()
```

# Informe Final: Identificación de la Tienda Menos Eficiente

## Introducción

El objetivo de este informe es analizar el rendimiento de cuatro tiendas en base a indicadores clave: ingresos totales, valoración promedio de los clientes, coste de envío promedio y precios promedio por categoría de producto. A partir de este análisis, se identificará cuál es la tienda menos eficiente, es decir, la que presenta mayores debilidades en relación con las demás.

---

## Análisis Comparativo

### 1. Ingresos Totales

- **Tienda 1**: \$1,150,880,400  
- **Tienda 2**: \$1,116,343,500  
- **Tienda 3**: \$1,098,019,600  
- **Tienda 4**: \$1,038,375,700  

> La **Tienda 4** presenta los ingresos totales más bajos, lo que sugiere menor volumen de ventas o menor capacidad de atracción de clientes.

---

### 2. Valoración Promedio de los Clientes

- **Tienda 1**: 3.98  
- **Tienda 2**: 4.04  
- **Tienda 3**: 4.05  
- **Tienda 4**: 4.00  

> Aunque las diferencias no son muy marcadas, la **Tienda 1** tiene la calificación más baja. Sin embargo, la Tienda 4 también se encuentra por debajo del promedio general.

---

### 3. Coste de Envío Promedio

- **Tienda 1**: \$26,018.61  
- **Tienda 2**: \$25,216.24  
- **Tienda 3**: \$24,805.68  
- **Tienda 4**: \$23,459.46  

> En este aspecto, la **Tienda 4** es la más eficiente, ya que ofrece el menor coste de envío. Este es un punto a favor dentro de su rendimiento general.

---

### 4. Precios Promedio por Categoría

#### Tienda 1:
- Electrodomésticos: \$1,165,657.69  
- Electrónicos: \$958,690.85  
- Promedio general: precios altos, especialmente en tecnología.

#### Tienda 2:
- Electrodomésticos: \$1,142,845.25  
- Electrónicos: \$973,533.41  
- Promedio: precios similares a Tienda 1, pero con menor dispersión.

#### Tienda 3:
- Electrodomésticos: **\$1,184,308.99** (el más alto)  
- Electrónicos: \$910,811.09  
- Promedio: precios elevados, especialmente en categorías clave.

#### Tienda 4:
- Electrodomésticos: \$1,115,197.64  
- Electrónicos: **\$907,929.27** (el más bajo)  
- Promedio: precios más bajos en varias categorías, lo que puede reducir el margen de ganancia.

> La **Tienda 4** presenta los precios más bajos en categorías importantes como electrónica y electrodomésticos, lo que podría afectar su rentabilidad si no está compensado por un volumen alto de ventas (que, como vimos, no es el caso).

---

## Conclusión

Tras evaluar todos los indicadores, se concluye que la **Tienda 4** es la **menos eficiente** entre las cuatro evaluadas.

### **Justificación:**
- Tiene el **nivel de ingresos más bajo**, lo que refleja bajo rendimiento comercial.
- Su **valoración de clientes** no se destaca frente a otras, permaneciendo por debajo del promedio.
- Aunque presenta el **coste de envío más bajo**, este no se traduce en mejores resultados de ventas.
- Tiene **precios más bajos en productos clave**, lo que puede indicar menor margen de beneficio y estrategia de precios poco competitiva a largo plazo.

En conjunto, estos factores sugieren que la Tienda 4 enfrenta mayores desafíos en términos de eficiencia y desempeño global, posicionándola como la **menos atractiva para una estrategia de ventas** como la que plantea el Sr. Juan.
